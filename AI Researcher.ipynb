```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
import os
from crewai import Crew , Task ,Agent
```

```python
load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')
model = ChatOpenAI(api_key=api_key,  model="gpt-3.5-turbo")
```

```python
model.invoke('hello there')
```

```python
research_agent = Agent(
    role = 'Researcher',
    goal = 'You will conduct a research about a given {topic}',
    backstory = """  
                You are a great researcher who conducts very technical research
                around a given {topic}. You research outcome contains facts only
                """ ,
    verbose = True,
    llm = model,
    allow_delegation = False,
)
```

```python
research_validator_agent = Agent(
    role = 'Research Validator',
    goal = 'Your job is to validate a given research paper',
    backstory = """ 
                You are a great or trusted research validator the outcome of your research
                is published once they are equivalent to industry standard. 
                """,
    
    verbose = True,
    llm = model,
    allow_delegation =True,
)
```

```python
social_media_post_agent = Agent(
    role = 'Blogger',
    goal = 'Your job is to write a blog post on a given research paper',
    backstory = """ 
                 Your job is to write a very friendly and technical blog post 
                 that can be posted on all social media platform from a given research paper.
                """,
    verbose = True,
    llm = model,
    allow_delegation = False,
)
```

```python
research_task  = Task(
    description = """ 
                    You stand out as an award winning researcher. You conduct very detail 
                    research based on a given {topic}. Your reseach papers are read by 
                    various people around the world because of how great they are and 
                    the reputation you have gathered over the years.
                """ ,
    expected_output = """ 
                     A very technical yet friendly research paper.
                    """,
    agent = research_agent,
    
)
```

```python
research_validator_task  = Task(
    description = """ 
                Being one of the greatest research paper validators in the world,
                your job is to validate research papers and make sure they are 
                only published after meeting the industry standards.
        """ ,
    expected_output = """ 
                    A great research paper ready for publication.
                    
                    """,
    agent =research_validator_agent,
 
)
```

```python
social_media_task  = Task(
    description = """ 
                    You're a social media content creator,you write blogs simplying 
                    research papers to your audience. Your priority is that it will very
                    friendly and easy to understand.
        """ ,
    expected_output = """ 
                    A well documented social media blog post not less that two pages.
                    
                    """,
    agent =social_media_post_agent,
 
)
```

```python
crew = Crew(
    agents = [research_agent,research_validator_agent,social_media_post_agent],
    tasks = [research_task,research_validator_task,social_media_task],
    memory = True ,
    verbose = 2
)
```

```python
inputs = {
    'topic' : 'Computer Vision'
}
```

```python
result = crew.kickoff(inputs = inputs)
```
